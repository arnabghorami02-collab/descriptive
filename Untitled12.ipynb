Q1: Understanding Central Tendency (Easy)

A bakery tracks the daily sales of muffins (in dozens) over a week: [10, 12, 11, 15, 14, 13, 12].
What is the most representative value of their weekly sales, and why?

Sol :     

Data: [10, 12, 11, 15, 14, 13, 12]

Mean = (10+12+11+15+14+13+12)/7 = 87/7 ≈ 12.43

Median = 12

Mode = 12

The most representative value is 12 (median/mode) because it reflects the central value without being affected by slightly higher numbers like 15.

Q4: Median in Real Life (Medium)

A car dealer notes the prices of used cars: [$8,000, $9,500, $10,200, $11,000, $50,000].

Why is the median a better measure than the mean in this case? Calculate the median.

Sol :      
Data: [$8000, $9500, $10200, $11000, $50000]

Median = $10,200

Median is better because $50,000 is an outlier that distorts the mean.
The median ($10,200) reflects the typical car price.

Q5: Dispersion Introduction (Medium)

A student times how long it takes to finish a puzzle each day: [25, 30, 27, 35, 40].
What does the range tell us about the variation in the student’s puzzle-solving time?

Sol :         

Data: [25, 30, 27, 35, 40]

Range = 40 − 25 = 15


 The range is 15 minutes, showing moderate variation in solving time.
